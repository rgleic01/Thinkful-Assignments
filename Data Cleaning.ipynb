```python
import csv
import numpy as np
import pandas as pd

#read csv file to Data Frame
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding = "ISO-8859-1")

#rename cost column
df.rename(index=str, columns={"COST (£) charged to Wellcome (inc VAT when charged)": "price"}, inplace=True)

#clean journal titles
def journal_clean(journal_string):
    journal_string = str(journal_string)
    #eliminate all numbers & stray punctuation
    for char in journal_string:
        if char.isalpha() == 0 and char != ' ': 
            journal_string = journal_string.replace(char, "")
    #make all characters lowercase
    journal_string = journal_string.lower()
    return journal_string
    
#apply title cleaning function
df['Journal title'] = df['Journal title'].apply(journal_clean)


#create a dictionary fixing values like 'americal' and linking them to the proper value 
#with more time for research, I could check to see all the variants and put them in this dictionary
cleaning_dict = {'current biol':'current biology', 
                 'americal' : 'american', 
                 'febs j': 'febs journal',
                 'heptology': 'hepatology',
                 'j biol chem': 'journal of biological chemistry',
                 'jounral': 'journal',
                 'jnl': 'journal',
                 ' org ': ' organic ',
                 'j med chem' : 'journal of medicinal chemistry'}


#function that applies dictionary to journal titles
def apply_dict(journal_title):
    if journal_title in cleaning_dict:
        return cleaning_dict.get(journal_title)
    else: return journal_title

#apply cleaning dictionary
df['Journal title'] = df['Journal title'].apply(apply_dict)


#print top 5 most published journals
df_top_five = (df.groupby('Journal title').count()).sort_values(by='price', ascending=False)
print(df_top_five.head())



#clean cost data
def price_clean(price_string):
    #remove currency symbols
    price_string = price_string.replace("$","").replace("£","")
    price_float = float(price_string)
    #remove 999999.00 values
    if price_float == 999999:
        price_float = None
    return price_float

#apply cleaning function to prices
df['price'] = df['price'].apply(price_clean)


#Calculate the mean, median, and standard deviation article cost for each journal.
df_mean = df.groupby('Journal title').mean()
df_median = df.groupby('Journal title').median()
df_std = df.groupby('Journal title').std()
print(df_median.head())
print(df_mean.head())
print(df_std.head())
```
