```python
import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
%matplotlib inline


exercise = sns.load_dataset("exercise")
print(exercise.head(n=5))

low_fat = exercise.loc[(exercise['diet'] == 'low fat')]
no_fat = exercise.loc[(exercise['diet'] == 'no fat')]
rest = exercise.loc[(exercise['kind'] == 'rest')]
walking = exercise.loc[(exercise['kind'] == 'walking')]
running = exercise.loc[(exercise['kind'] == 'running')]


#histogram for two diets
plt.hist(low_fat['pulse'], color='red', alpha=0.5, label='Low Fat Diet')
plt.hist(no_fat['pulse'], color='blue', alpha=0.5, label='No Fat Diet')
plt.title('Plot 1: Pulse Rate for Different Diets')
plt.xlabel('Pulse')
plt.legend(loc='upper right')
plt.show()

#histogram for different activities
plt.hist(rest['pulse'], color='red', alpha=0.5, label='Rest')
plt.hist(walking['pulse'], color='blue', alpha=0.5, label='Walking')
plt.hist(running['pulse'], color='green', alpha=0.5, label='Running')
plt.title('Plot 2: Pulse Rate for Different Activities')
plt.xlabel('Pulse')
plt.legend(loc='upper right')
plt.show()



diet_box = sns.boxplot(x = exercise['pulse'], y = exercise['diet'])
plt.title('Plot 3: Pulse Rate by Diet')
plt.show()

activity_box = sns.boxplot(x = exercise['pulse'], y = exercise['kind'])
plt.title('Plot 4: Pulse Rate by Diet')
plt.show()
```

This data set shows the distribution in heart rate based on the diet (low fat vs. no fat), type of activity (rest, walking, or running), and duration of activity (1, 15, or 30 minutes).

There are a few things to notice from these graphs. The histogram shows that no fat and low fat diets yielded similar heart rates, but the box plot showed that a no fat diet produced a higher median heart rate and larger distribution. And running unsurprisingly produced a wide range of heart rates due to the huge difference between running for 1 minute vs running for 30 minutes.
