```python
#specify the longtitude and latitude 

url='https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=38.8976763%2C-77.0365298&radius=1000&keyword=restaurant&key={}'.format(apiKey)
response = requests.get(url)
resp_json= response.json()
```

```python
resp_json['results'][0]['name'], resp_json['results'][0]['vicinity'], resp_json['results'][0]['rating']
```

```python
names=[]
locations=[] 
ratings=[] 
for i in range(20): 
    names.append(resp_json['results'][i]['name'])
    locations.append(resp_json['results'][i]['vicinity'])
    ratings.append(resp_json['results'][i]['rating'])
```

```python
import pandas as pd 
df=pd.DataFrame()
df['name']=names 
df['location']=locations 
df['rating']=ratings 
```

```python
df.to_csv('Restaurants_Washington.csv', index=False)
```
